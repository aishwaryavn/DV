Box Plot  shows 5 statistically significant numbers- the minimum, the 25th percentile, the median, the 75th percentile and the maximum. It is thus useful for visualizing the spread of the data is and deriving inferences accordingly. 

```r
#Single box plot
boxplot(airquality$Solar.R)
```

# Histogram
Histogram is basically a plot that breaks the data into bins (or breaks) and shows frequency distribution of these bins.

```r
# 4 figures arranged in 2 rows and 2 columns
attach(mtcars)
par(mfrow=c(2,2))
plot(wt,mpg, main="Scatterplot of wt vs. mpg")
plot(wt,disp, main="Scatterplot of wt vs disp")
hist(wt, main="Histogram of wt")
boxplot(wt, main="Boxplot of wt")
```

```r
# 3 figures arranged in 3 rows and 1 column
attach(mtcars)
par(mfrow=c(3,1))
hist(wt)
hist(mpg)
hist(disp)
```

```r
data(iris)
par(mfrow=c(2,2))
boxplot(iris$Sepal.Length,col="red")
boxplot(iris$Sepal.Length~iris$Species,col="red")
boxplot(iris$Sepal.Length~iris$Species,col=heat.colors(3))
boxplot(iris$Sepal.Length~iris$Species,col=topo.colors(3))
```

```r
n=seq(1:12)
barplot(n)
```

```r
n=12
pie(rep(1,n), col=FALSE)
```

```r
pie(rep(1,n), col=rainbow(n))
```

```r
#If you would prefer something skewed to the reds and yellows, use heat.

pie(rep(1,n), col=heat.colors(n))
```

```r
#If you prefer something in earth tones, use terrain.

pie(rep(1,n), col=terrain.colors(n))
```

```r
#If you prefer these this type of palette, but with a bit more blue, try topo.

pie(rep(1,n), col=topo.colors(n))
```

```r
#Finally, if you prefer subdued pinks and blues, try cm.
pie(rep(1,n), col=cm.colors(n))
```
