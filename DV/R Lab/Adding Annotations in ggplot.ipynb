```r
# library
library(ggplot2)
 
# Keep 30 first rows in the mtcars natively available dataset
data = head(mtcars, 30)
 
# 1/ add text with geom_text, use nudge to nudge the text
```

```r
ggplot(data, aes(x=wt, y=mpg)) +
  geom_point() +  # Show dots
  geom_text(
    label=rownames(data), 
    nudge_x = 0.2, nudge_y = 0.2, 
    check_overlap = T, angle = 10, size = 4
  )
```

```r
ggplot(data, aes(x=wt, y=mpg)) +
  geom_point() + # Show dots
  geom_label(
    label=rownames(data), 
    nudge_x = 0.25, nudge_y = 0.25, 
    check_overlap = T
  )
```

```r
ggplot(data, aes(x=wt, y=mpg)) +
  geom_point() + # Show dots
  geom_label(
    label="Look at this!", 
    x=4,
    y=30,
label.padding = unit(0.2, "lines"), # Rectangle size around label
    label.size = 15,
  label.r = unit(0.10, "lines"),
    color = "green",
    fill="#69b3a2", fontface="bold"
  ) +
annotate("rect", xmin=c(2,4), xmax=c(3,5), ymin=c(20,10) , ymax=c(30,20), alpha=0.2, color="red", fill="blue") +
annotate("segment", x = 1, xend = 3, y = 25, yend = 15, colour = "purple", size=5, alpha=0.6) +
annotate("segment", x = 2, xend = 4, y = 15, yend = 25, colour = "pink", size=3, alpha=0.6, arrow=arrow())
```

```r
library(dplyr)
library(tibble)
data=head(mtcars, 30)

# Change data rownames as a real column called 'carName'
data <- data %>%
  rownames_to_column(var="carName")
  
data
```

```r
# Plot
ggplot(data, aes(x=wt, y=mpg)) +
  geom_point() + 
  geom_label( 
    data=data %>% filter(mpg>20 & wt>3), # Filter data first
    aes(label=carName)
  )
```
