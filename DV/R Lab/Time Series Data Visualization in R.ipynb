```r
# Basic line plot
ggplot(data = economics, aes(x = date, y = pop))+
  geom_line(color = "#00AFBB", size = 2)
# Plot a subset of the data
ss <- subset(economics, date > as.Date("2006-1-1"))
ggplot(data = ss, aes(x = date, y = pop)) + 
  geom_line(color = "#FC4E07", size = 2)
```

```r
library(tidyr)
library(dplyr)
df <- economics %>%
  select(date, psavert, uempmed) %>%
  gather(key = "variable", value = "value", -date)
head(df, 10)
```

```r
ggplot(df, aes(x = date, y = value)) + 
  geom_line(aes(color = variable), size = 1) +
  scale_color_manual(values = c("#00AFBB", "#E7B800")) +
  theme_test()
# theme_gray, theme_bw, theme_linedraw, theme_light, theme_dark, theme_minimal, theme_classic, theme_void, theme_test
```

```r
ggplot(df, aes(x = date, y = value)) + 
  geom_area(aes(color = variable, fill = variable), 
            alpha = 0.5, position = position_dodge(0.8)) +
  scale_color_manual(values = c("#00AFBB", "#E7B800")) +
  scale_fill_manual(values = c("#00AFBB", "#E7B800"))
```

```r
# Base plot with date axis
p <- ggplot(data = economics, aes(x = date, y = psavert)) + 
     geom_line(color = "#00AFBB", size = 1)
p
# Set axis limits c(min, max)
min <- as.Date("2002-1-1")
max <- NA
p + scale_x_date(limits = c(min, max))
```

```r
p + stat_smooth(
  color = "#FC4E07", fill = "#FC4E07",
  method = "loess"
  )
```

```r
# Format : month/year
p + scale_x_date(date_labels = "%b/%Y")  + xlab("")
p + scale_x_date(date_labels = "%m/%d")
p + scale_x_date(date_labels = "%B-%d")

#%d	Day of the month (decimal number)
#%m	Month (decimal number)
#%b	Month (abbreviated)
#%B	Month (full name)
#%y	Year (2 digit)
#%Y	Year (4 digit)
```

```r
p + theme(axis.text.x=element_text(angle=60, hjust=1, colour = "#FC4E07")) +
geom_hline(yintercept=10, color="orange", size=.5) 

```
